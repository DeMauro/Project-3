### Explore & Clean Data

```python
import pandas as pd

path= "../Resources/loan_data.json"
df= pd.read_json(path)
df
```

```python
# Remove nulls
clean_df= df.dropna()
clean_df
```

### Transform Data

```python
# Change Male, Married, Graduate, Self_Employeed & Approved Application Status= 1 
one_df=clean_df.replace(["Male", "Yes", "Graduate", "Y"],1)

# Change Female, NOT Married, NOT Graduate, NOT Self_Employeed & Decline Application Status= 0 
zero_df=one_df.replace(["Female","No", "Not Graduate", "N"],0)

# Let us view what we got
zero_df
```

```python
# Explore the Property Area Column 
zero_df["Property_Area"].value_counts()
```

```python
# Explore the Income Column 
zero_df["Income"].value_counts()
```

```python
# Convert Property Area Column
# Semiurban, low= 0
semi_df=zero_df.replace(["Semiurban", "low"],0)

# Urban, medium= 1
urban_df=semi_df.replace(["Urban", "medium"],1)

# Rural, high= 2
rural_df=urban_df.replace(["Rural", "high"],2)

rural_df
```

```python
# Fixing Dependents of 3+ to 3
final_df=rural_df.replace("3+",3)
final_df
```

```python
# not sure what credit history lebling means.
# I am assuming 1=yes and 2=no
df["Credit_History"].value_counts()
```

### Create Variables

```python
# Create Variables
X=final_df.drop(["Application_ID","Application_Status"], axis=1)
y=final_df["Application_Status"]

print(X.shape, y.shape)
```

### Test - Train - Split

```python
#Test, Train, Split Data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# Create a decision Tree
from sklearn import tree

# Create a Model
model=tree.DecisionTreeClassifier()

#Fit the model
model.fit(X_train,y_train)
```

```python
# checking the quality score
model.score(X_test, y_test)
```

```python
model.predict([[0,1,1,0,1,0,1,0]])
```
