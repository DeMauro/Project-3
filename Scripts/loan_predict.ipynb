### Load & Join Data


```python
# Loading in data

import pandas as pd
import numpy as np
import pandas as pd

import warnings
warnings.simplefilter('ignore', FutureWarning)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical



info1=pd.read_csv("../Resources/loan_table.csv")

info1
```

```python
info2=pd.read_csv("../Resources/borrower_table.csv")

info2
```

```python
new_df=info1.set_index("loan_id").join(info2.set_index("loan_id"))
join_df=new_df.drop("date", axis=1)
join_df
```

### Trasform Data

```python
join_df[['loan_repaid','fully_repaid_previous_loans','currently_repaying_other_loans']] = join_df[['loan_repaid','fully_repaid_previous_loans','currently_repaying_other_loans']].fillna(2)
join_df
```

```python
join_df['avg_percentage_credit_card_limit_used_last_year'] = join_df['avg_percentage_credit_card_limit_used_last_year'].fillna(0)
join_df
```

```python
join_df["loan_purpose"].value_counts()
```

```python
home_df=join_df.replace("home",0)
business_df=home_df.replace("business",1)
investment_df=business_df.replace("investment",2)
ef_df=investment_df.replace("emergency_funds",3)
repaid_df=ef_df.replace("other",4)
```

```python
final_df =repaid_df.drop("loan_repaid", axis=1)
final_df.head()
```

```python
repaid_df.head()
```

```python
repaid_df.to_csv('../Resources/outcome_df.csv', index=False)
```

### Test-Train-Split


```python
X=final_df.drop("loan_granted", axis=1)
y=final_df["loan_granted"]

print(X.shape, y.shape)
```

```python
#Test, Train, Split Data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

### Decision Tree

```python
# Create a decision Tree
from sklearn import tree

# Create a Model
model=tree.DecisionTreeClassifier()

#Fit the model
model.fit(X_train,y_train)
```

```python
# checking the quality score
model.score(X_test, y_test)
```
